```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import scipy

px, dt = np.meshgrid(pixelsize, dwelltime)
xs = px
ys = dt
zs = optimal_phase_delay

# plot raw data
plt.figure()
ax = plt.subplot(111, projection='3d')
ax.scatter(xs, ys, zs, color='b')
ax.view_init(20, 130)
ax.set_zlim(0,140)

points = []
for idxx, px in enumerate(pixelsize):
    for idxy, py in enumerate(dwelltime):
        points.append([px,py,optimal_phase_delay[idxy,idxx]])
data = np.array(points)
        
# best-fit linear plane
A = np.c_[data[:,0], data[:,1], np.ones(data.shape[0])]
C,_,_,_ = scipy.linalg.lstsq(A, data[:,2])    # coefficients

# evaluate it on grid
X,Y = np.meshgrid(np.arange(0, 0.2, 0.01), np.arange(0, 0.5, 0.01))
Z = C[0]*X + C[1]*Y + C[2]
Z[Z<0] = 0

# plot points and fitted surface
fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, alpha=0.2)
ax.scatter(data[:,0], data[:,1], data[:,2], c='r', s=50)
plt.xlabel('X')
plt.ylabel('Y')
ax.set_zlabel('Z')
#ax.axis('equal')
ax.axis('tight')
ax.view_init(20, 130)
ax.set_zlim(0,140)
plt.show()
```
