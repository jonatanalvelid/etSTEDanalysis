# Save etSTED data area summary from one experiment run, with all events and an overview widefield image

```python
import os
import glob
import numpy as np
import tifffile
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from scipy import ndimage as ndi
```

```python
folder_top = 'example_data'
dirs_top = []
for item in os.listdir(folder_top):
    if os.path.isdir(os.path.join(folder_top, item)) and 'nontriggered' not in item and 'false' not in item:
        dirs_top.append(os.path.join(folder_top, item))
print(dirs_top)

save = True

for folder in dirs_top:
    area_name = folder.split('\\')[-1]
    dirs = [x[0] for x in os.walk(folder)][1:]
    dirs = [dir for dir in dirs if 'false' not in dir]

    if 'stedresolution' in folder or 'stedtimelapse' or 'triggeredtimelapses' in folder:

        for file in os.listdir(folder):
            if file.endswith('raw.tif'):
                wfimgfile = file
                break

        wfimg = tifffile.imread(os.path.join(folder, wfimgfile))
        wfimg_flat = np.sum(wfimg,0)

        stedfiles = []
        for file in os.listdir(folder):
            if file.endswith('STED.tif'):
                stedfiles.append(os.path.join(folder, file))

        wffiles = []
        for file in os.listdir(folder):
            if file.endswith('raw.tif'):
                wffiles.append(os.path.join(folder, file))

        noise_level = 250

        # create figure
        fig1, axs = plt.subplots(len(dirs), 4, figsize=(30,7*len(dirs)))
        #fig1, axs = plt.subplots(3, 4, figsize=(30,7*3))

        for idx, recdir in enumerate(dirs):
            #if idx < 3:
            if 'false' not in recdir and '\false' not in recdir and 'rest' not in recdir:
                logfiles = []

                # get logfile
                for file in os.listdir(recdir):
                    if file.endswith('log.txt'):
                        logfile = os.path.join(recdir, file)
                        logfiles.append(logfile)
                        break

                # read triggered coordinate
                f0 = open(logfile, 'r')
                lines = f0.readlines()
                for line in lines:
                    param = line.split(' ')[0]
                    if param == 'fastscan_x_center:':
                        xcenter = int(line.split(' ')[1].split('\n')[0])
                    elif param == 'fastscan_y_center:':
                        ycenter = int(line.split(' ')[1].split('\n')[0])
                    elif param == 'pipeline_start':
                        t_str = line.split(' ')[1].split('\n')[0]
                        t_s = float(t_str.split('s')[0])
                        t_us = 1e-6 * float(t_str.split('s')[1].split('us')[0])
                        t_pipelinestart = t_s + t_us
                    elif param == 'scan_start' or param == 'scan_start_frame0':
                        t_str = line.split(' ')[1].split('\n')[0]
                        t_s = float(time_str.split('s')[0])
                        t_us = 1e-6 * float(time_str.split('s')[1].split('us')[0])
                        t_scanstart = t_s + t_us
                        t_wftoscan = (t_scanstart - t_pipelinestart) * 1e3  # ms
                    elif param == 'scan_start_frame1':
                        t_str = line.split(' ')[1].split('\n')[0]
                        t_s = float(time_str.split('s')[0])
                        t_us = 1e-6 * float(time_str.split('s')[1].split('us')[0])
                        t_scanframe1 = t_s + t_us
                        t_scanperiod = (t_scanframe1 - t_scanstart) * 1e3  # ms
                    elif param == 'pipeline_end':
                        t_str = line.split(' ')[1].split('\n')[0]
                        t_s = float(t_str.split('s')[0])
                        t_us = 1e-6 * float(t_str.split('s')[1].split('us')[0])
                        t_pipelineend = t_s + t_us
                        t_pipeline = (t_pipelineend - t_pipelinestart) * 1e3  # ms
                coord_temp = np.array([xcenter, ycenter])
                if idx==0:
                    coords = coord_temp
                else:
                    coords = np.vstack((coords,coord_temp))

            # get sted stack/image and wf stack
            stedstack = tifffile.imread(stedfiles[idx])
            if stedstack.ndim == 3:
                stedimg = np.copy(stedstack[0])
            else:
                stedimg = stedstack
            wfstack = tifffile.imread(wffiles[idx])
            wfimg = np.copy(wfstack[-1])

            # create preprocessed image
            anaimg = np.subtract(wfstack[-1], wfstack[-2])
            divimg = np.copy(wfstack[-2])
            divimg[divimg < noise_level] = 100000
            anaimg = np.divide(anaimg, divimg)
            anaimg = ndi.filters.gaussian_filter(anaimg, 2)
            anaimg[anaimg < 0] = 0

            # get calcium signal in surrouding area
            radius = 7
            lim_signal = 0.4
            eventstack = np.copy(wfstack[:,coord_temp[1]-radius:coord_temp[1]+radius,coord_temp[0]-radius:coord_temp[0]+radius])
            #print(eventstack)
            event_signal = np.array([slice.sum() for slice in eventstack])
            #print(event_signal)
            event_signal = event_signal / np.max(event_signal)
            event_signal = event_signal[event_signal > lim_signal]

            # plot images and calcium signal
            axs[idx,0].imshow(wfimg, vmax=np.max(np.max(wfimg))/1.5)
            axs[idx,0].scatter(coord_temp[0], coord_temp[1], s=200, facecolors='none', edgecolors='red')
            axs[idx,0].tick_params(axis='both', which='both', bottom=False, top=False,  left=False, right=False, labelleft=False, labelbottom=False)
            axs[idx,0].title.set_text('WF triggering frame')
            axs[idx,1].imshow(anaimg, vmax=np.max(np.max(anaimg)), cmap=plt.get_cmap('gray'))
            axs[idx,1].scatter(coord_temp[0], coord_temp[1], s=200, facecolors='none', edgecolors='red')
            axs[idx,1].tick_params(axis='both', which='both', bottom=False, top=False,  left=False, right=False, labelleft=False, labelbottom=False)
            axs[idx,1].title.set_text('Preprocessed frame')
            stedplot = axs[idx,2].imshow(stedimg, vmax=np.max(np.max(stedimg)), cmap=plt.get_cmap('hot'))
            axs[idx,2].tick_params(axis='both', which='both', bottom=False, top=False,  left=False, right=False, labelleft=False, labelbottom=False)
            cb = fig1.colorbar(stedplot, ax=axs[idx,2])
            axs[idx,2].title.set_text('Scanned STED image')
            #plt.setp(cb.ax.get_yticklabels(), visible=False)
            axs[idx,3].plot(event_signal)
            axs[idx,3].set_xlabel('WF frame')
            axs[idx,3].set_ylabel('Summed BAPTA signal [norm. int]')
            #axs[idx,4].imshow(eventstack[4], vmax=np.max(np.max(eventstack[-1])))

        fig2 = plt.figure(figsize=(30,30))
        plt.imshow(wfimg_flat, vmax=np.max(np.max(wfimg_flat))/1.5)
        if len(coords)==2:
            plt.scatter(coords[0], coords[1], s=600, facecolors='none', edgecolors='red')
        else:
            for row in coords:
                plt.scatter(row[0], row[1], s=600, facecolors='none', edgecolors='red')
        plt.tick_params(axis='both', which='both', bottom=False, top=False,  left=False, right=False, labelleft=False, labelbottom=False)
        fig2.suptitle('All detected events in area', fontsize=20)

        # save summary to pdf
        if save:
            pdf = PdfPages(os.path.join(folder_top,f'area_summary_{area_name}.pdf'))
            pdf.savefig(fig1)
            pdf.savefig(fig2)
            pdf.close()

    elif 'validationruns' in folder: 

        for file in os.listdir(folder):
            if file.endswith('raw.tif'):
                wfimgfile = file
                break

        wfimg = tifffile.imread(os.path.join(folder, wfimgfile))
        wfimg_flat = np.sum(wfimg,0)

        rawfiles = []
        for file in os.listdir(folder):
            if file.endswith('raw.tif'):
                rawfiles.append(os.path.join(folder, file))

        anafiles = []
        for file in os.listdir(folder):
            if file.endswith('ana.tif'):
                anafiles.append(os.path.join(folder, file))

        # create figure
        fig1, axs = plt.subplots(len(dirs), 3, figsize=(25,8*len(dirs)))
        #fig1, axs = plt.subplots(3, 3, figsize=(25,8*3))

        for idx, recdir in enumerate(dirs):
            #if idx < 3:
            if 'false' not in recdir and '\false' not in recdir and 'rest' not in recdir:
                logfiles = []

                for file in os.listdir(recdir):
                    if file.endswith('log.txt'):
                        logfile = os.path.join(recdir, file)
                        logfiles.append(logfile)
                        break

                f0 = open(logfile, 'r')
                lines = f0.readlines()
                for line in lines:
                    param = line.split(' ')[0]
                    if param == 'fastscan_x_center:':
                        xcenter = int(line.split(' ')[1].split('\n')[0])
                    elif param == 'fastscan_y_center:':
                        ycenter = int(line.split(' ')[1].split('\n')[0])
                coord_temp = np.array([xcenter, ycenter])
                if idx==0:
                    coords = coord_temp
                else:
                    coords = np.vstack((coords,coord_temp))

            rawstack = tifffile.imread(rawfiles[idx])
            rawimg = np.copy(rawstack[3])
            anastack = tifffile.imread(anafiles[idx])
            anaimg = np.copy(anastack[3])

            radius = 7
            lim_signal = 0.4
            eventstack = np.copy(rawstack[:,coord_temp[1]-radius:coord_temp[1]+radius,coord_temp[0]-radius:coord_temp[0]+radius])
            #print(eventstack)
            event_signal = np.array([slice.sum() for slice in eventstack])
            #print(event_signal)
            event_signal = event_signal / np.max(event_signal)
            event_signal = event_signal[event_signal > lim_signal]

            axs[idx,0].imshow(rawimg, vmax=np.max(np.max(rawimg))/1.5)
            axs[idx,0].scatter(coord_temp[0], coord_temp[1], s=200, facecolors='none', edgecolors='red')
            axs[idx,0].tick_params(axis='both', which='both', bottom=False, top=False,  left=False, right=False, labelleft=False, labelbottom=False)
            axs[idx,0].title.set_text('WF triggering frame')
            axs[idx,1].imshow(anaimg, vmax=np.max(np.max(anaimg)), cmap=plt.get_cmap('gray'))
            axs[idx,1].scatter(coord_temp[0], coord_temp[1], s=200, facecolors='none', edgecolors='red')
            axs[idx,1].tick_params(axis='both', which='both', bottom=False, top=False,  left=False, right=False, labelleft=False, labelbottom=False)
            axs[idx,1].title.set_text('Preprocessed frame')
            axs[idx,2].plot(event_signal)
            axs[idx,2].set_xlabel('WF frame')
            axs[idx,2].set_ylabel('Summed BAPTA signal [norm. int]')
            #axs[idx,4].imshow(eventstack[4], vmax=np.max(np.max(eventstack[-1])))

        fig2 = plt.figure(figsize=(25,25))
        plt.imshow(wfimg_flat, vmax=np.max(np.max(wfimg_flat))/1.5)
        if len(coords)==2:
            plt.scatter(coords[0], coords[1], s=600, facecolors='none', edgecolors='red')
        else:
            for row in coords:
                plt.scatter(row[1], row[0], s=600, facecolors='none', edgecolors='red')
        plt.tick_params(axis='both', which='both', bottom=False, top=False,  left=False, right=False, labelleft=False, labelbottom=False)
        fig2.suptitle('All detected events in area', fontsize=20)

        # save summary to pdf
        if save:
            pdf = PdfPages(os.path.join(folder_top,f'area_summary_{area_name}.pdf'))
            pdf.savefig(fig1)
            pdf.savefig(fig2)
            pdf.close()    
```
